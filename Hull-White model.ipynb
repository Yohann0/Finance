# Hull White Model Theory

We intend to value interest rate derivatives using the **Hull White Model**. The evolution of the short rate process is determined as follows:

$$
d r_{t}=\left(\theta(t)-a r_{t}\right) d t+\sigma d W_{t} \tag{1}
$$

The introduction of a time-variant $\theta$ aims to perfectly fit the initial term structure. Let $f^{M}(0, T)$ denote the observed instantaneous forward rate at the current time:

$$
f^{M}(0, t)=-\frac{\partial \ln P^{M}(0, T)}{\partial T} \tag{2}
$$

The HW model reproduces the current yield curve if:

$$
\theta(t)=\frac{\partial f^{M}}{\partial T}(0, T)+a f^{M}(0, T)+\frac{\sigma^{2}}{2 a}\left(1-e^{-2 a t}\right) \tag{3}
$$

Assuming the previous condition is fulfilled, we have:

$$
r_{t}=r_{s} e^{-a(t-s)}+\alpha(t)-\alpha(s) e^{-a(t-s)}+\sigma \int_{s}^{t} e^{-a(t-u)} d W_{u} \tag{4}
$$

$$
\alpha(t)=f^{M}(0, T)+\frac{\sigma^{2}}{2 a^{2}}\left(1-e^{-a t}\right)^{2} \tag{5}
$$

$$
\mathbb{E}\left[r_{t} \mid \mathcal{F}_{s}\right]=r_{s} e^{-a(t-s)}+\alpha(t)-\alpha(s) e^{-a(t-s)} \tag{6}
$$

$$
\operatorname{Var}\left[r_{t} \mid \mathcal{F}_{s}\right]=\frac{\sigma^{2}}{2 a}\left(1-e^{-2 a(t-s)}\right) \tag{7}
$$

The price of the zero-coupon bond is given by:

$$
P(t, T)=A(t, T) e^{-\Phi(t, T) r(t)} \tag{8}
$$

$$
A(t, T)=\frac{P^{M}(0, T)}{P^{M}(0, t)} \exp \left[\Phi(t, T) f^{M}(0, T)-\frac{\sigma^{2}}{4 a}\left(1-e^{-2 a t}\right) \Phi^{2}(t, T)\right] \tag{9}
$$

$$
\Phi(t, T)=\frac{1}{a}\left(1-e^{-a(T-t)}\right) \tag{10}
$$

#### 6. What is the economic foundation behind mean reversion in interest rate models?

Mean reversion in interest rate models is based on the economic principle that interest rates tend to move towards an average level over time. Over the long term, the economic factors influencing interest rates tend to stabilize around certain levels, leading interest rates to revert to an equilibrium level. In efficient markets, if rates deviate significantly from historical averages, market participants may anticipate a correction, leading to mean reversion. Central banks also play a role in maintaining interest rates around their target levels.

#### 7. Define the **T-forward measure** and derive the Hull and White Model dynamics under this measure.

The T-forward measure is the measure associated with the numeraire which is the zero-coupon bond maturing at time T. 

We note $B_t$ the money market account, and $N_t$ the numeraire of the T-forward measure. We have:
$$dB_t = r_t B_t dt$$
$$dN_t = r_t N_t dt + \tilde \sigma_t N_t dW_t$$
where $\tilde \sigma_t$ is the volatility of the numeraire.

Applying Itô's lemma to $\frac{1}{N_t}$, we get:
$$\begin{align*}
d\left(\frac{1}{N_t}\right) &= -\frac{1}{N_t^2} dN_t + \frac{1}{N_t^3} (dN_t)^2 \\
&= -\frac{1}{N_t} (r_t dt + \tilde \sigma_t dW_t) + \frac{1}{N_t} \tilde \sigma_t^2 dt
\end{align*}$$

Then, we have:
$$\begin{align*}
d\left(\frac{B_t}{N_t}\right) &= B_t d\left(\frac{1}{N_t}\right) + \frac{1}{N_t} dB_t \\
&= \frac{B_t}{N_t} \left[ (-r_t + \tilde \sigma_t^2) dt - \tilde \sigma_t dW_t \right] + \frac{B_t}{N_t} r_t dt \\
&= \frac{B_t}{N_t} \left( \tilde \sigma_t^2 dt - \tilde \sigma_t dW_t \right)
\end{align*}$$

$\frac{B_t}{N_t}$ is a martingale under the T-forward measure, and because the diffusion is term is constant through a change of measure, we have:
$$d\left(\frac{B_t}{N_t}\right) = - \tilde \sigma_t \frac{B_t}{N_t} d \tilde W_t$$
where $\tilde W_t$ is a Brownian motion under the T-forward measure.

By identification, we find:
$$d \tilde W_t = dW_t - \tilde \sigma_t dt$$

By using equation (6), we have:
$$N_t = P(t,T) = A(t,T) e^{-\Phi(t,T) r_t}$$

By applying Itô's lemma to $N_t$, we get:
$$\begin{align*}
dN_t &= [...] dt + A(t,T) (- \Phi(t,T)) e^{-\Phi(t,T) r_t} dr_t \\
&= [...] dt - \sigma \Phi(t,T) N_t dW_t
\end{align*}$$

By identification, we find:
$$\tilde \sigma_t = - \sigma \Phi(t,T)$$

Finally, we have:
$$d\tilde W_t = dW_t + \sigma \Phi(t,T) dt $$
and
$$dr_t = (\theta (t) - \sigma² \Phi(t,T) - a r_t) dt + \sigma d \tilde W_t$$

This the dynamics of the Hull-White model under the T-forward measure.

#### 8. Given the term structure of interest rates, construct a swap pricer that calculates the forward swap interest rate for an *in fine* bond.

```python
t = 0 #start
n = 6 #nb periods
delta = 1/2 #frequency
N = 1e6 #nominal

tau, beta0, beta1, beta2 = params

def get_swap_rate(delta, t, n, tau, beta0, beta1, beta2):
    payments= np.linspace(t, t+n*delta, n+1)
    model_yields = Nelson_Siegel(tau, beta0, beta1, beta2, payments)
    bonds = np.exp(-model_yields*payments)
    bonds[0]=1
    def VAN_fixed(bonds, S, delta):
        return S*delta*np.sum(bonds[1:])
    def VAN_float(bonds):
        return np.sum([bonds[i-1]-bonds[i] for i in range(1, len(bonds))])
    def error(S):
        return (VAN_fixed(bonds, S, delta)-VAN_float(bonds))**2

    return sc.optimize.fmin(error, 0.01)[0]

S = get_swap_rate(delta, t, n, tau, beta0, beta1, beta2)
print("Swap rate: {:.2%}".format(S))
```

#### 9. Compare the obtained valuation with the theoretical value (using the exact formula for the swap rate).

The forward swap rate between $T_\alpha$ and $T_\beta$ is given by
$$
S_{\alpha\beta}(t) = \frac{P(t, T_\alpha) - P(t, T_\beta)}{\sum_{i=\alpha+1}^\beta \tau_i P(t, T_i)}
$$
where $\tau_i = T_i - T_{i-1}$ is the time between two consecutive cash flows.

We can rewrite it as 
$$
S_{\alpha\beta} = \frac{1-\prod_{j=\alpha+1}^\beta \frac{1}{1+\tau_j F_j(t)}}{\sum_{i=\alpha+1}^\beta \tau_i \prod_{j=\alpha+1}^\beta \frac{1}{1+\tau_j F_j(t)}}
$$

```python
def get_swap_rate_formula(delta, t, n, tau, beta0, beta1, beta2):
    payments= np.linspace(t, t+n*delta, n+1)
    model_yields = Nelson_Siegel(tau, beta0, beta1, beta2, payments)
    alpha=0
    beta=payments.shape[0]-1
    bonds = np.exp(-model_yields * payments)
    bonds[0] = 1
    S = (bonds[alpha] - bonds[beta]) / (delta*np.sum(bonds[alpha+1:beta+1]))
    return S

S_formula = get_swap_rate_formula(delta, t, n, tau, beta0, beta1, beta2)
print(f"Swap rate using the formula: {S_formula:.2%}")
```

We are quite close to the swap rate we computed with the minimizer.

#### 10. How to extend the swap pricer to an amortizing loan?

To extend the swap pricer to handle amortizing loans, we need to account for the intermediate cash flows and the decreasing notional amount over the life of the swap. 

```python
def get_swap_rate_amortization(delta, t, n, tau, beta0, beta1, beta2, amortization_schedule):
    payments= np.linspace(t, t+n*delta, n+1)
    model_yields = Nelson_Siegel(tau, beta0, beta1, beta2, payments)
    bonds = np.exp(-model_yields * payments)
    bonds[0] = 1

    def VAN_fixed(bonds, S, delta, amortization_schedule):
        return np.sum([S * delta * amortization_schedule[i-1] * bonds[i] for i in range(1, len(bonds))])

    def VAN_float(bonds, amortization_schedule):
        return np.sum([amortization_schedule[i-1] * (bonds[i-1] - bonds[i]) for i in range(1, len(bonds))])

    def error(S):
        return (VAN_fixed(bonds, S, delta, amortization_schedule) - VAN_float(bonds, amortization_schedule))**2

    return sc.optimize.fmin(error, 0.01)[0]

# Linear amortization schedule
amortization_schedule = np.linspace(N, 0, n + 1)

S = get_swap_rate_amortization(delta, t, n, tau, beta0, beta1, beta2, amortization_schedule)
print("Swap rate: {:.2%}".format(S))
```

We find a swap rate lower than without amortizing which is coherent with the upward trend of the yield curve because less nominal is applied on higher long-term rates.

#### 11. Determine the cap value with a Monte Carlo simulation under the risk-neutral measure.

```python
# Parameters definition
r0 = .03
a = .1
sigma = .01
T = 10
delta=1
K = .02
t=0
NT = 100
dt = T/NT
N=5000
```

```python
def theta(t, a = a, beta0 = beta0, beta1 = beta1, beta2 = beta2, tau = tau, sigma=sigma):
     return -beta1 / tau * np.exp(-t / tau) + beta2 / tau * np.exp(-t / tau) - beta2 * t / tau**2 * np.exp(-t / tau) + a * (beta0 + beta1 * np.exp(-t / tau) + beta2 * t / tau * np.exp(-t / tau)) + sigma**2 / (2 * a) * (1 - np.exp(-2 * a * t))

np.random.seed(123)
def mc_sim_risk_neutral_measure(T, N, r0, a, sigma, theta, dt):
    r = np.zeros((NT+1,N))
    r[0,:] = r0
    for i in range(NT):
        eps = np.random.normal(0,1,N)
        r[i+1,:] = r[i,:]+(theta(i*dt)-a*r[i,:])*dt+sigma*np.sqrt(dt)*eps
    return r

r_rn = mc_sim_risk_neutral_measure(NT, N, r0, a, sigma, theta, dt)
plt.plot(r_rn[:, :1000])
plt.show()
```

```python
def P(t,T,r):
    """ZCB price"""
    return A(t,T)*np.exp(-Phi(t,T)*r[int(t/dt),:])
def A(t,T):
    return P_M(T)/P_M(t)*np.exp(Phi(t,T)*f_M(T)-sigma**2/(4*a)*(1-np.exp(-2*a*t))*Phi(t,T)**2)
def Phi(t,T):
    return (1-np.exp(-a*(T-t)))/a
def P_M(T):
    integral, _ = sc.integrate.quad(f_M, 0, T)
    return np.exp(-integral)
    # return np.exp(-Nelson_Siegel(tau,beta0,beta1,beta2,T)*T)
    
def f_M(t):
    return Nelson_Siegel(tau, beta0, beta1, beta2, t)
    
def cap_price_risk_neutral(t, T, delta, K, r):
    price=0
    for ti in np.arange(t, T, delta):
        T1 = ti
        T2 = ti+delta
        P_T1_T2 = P(T1,T2,r)
        libor = (1-P_T1_T2)/(delta*P_T1_T2)
        payoff = np.maximum(libor-K,0)*delta
        discounted_payoff = payoff*P_T1_T2
        caplet_price = np.mean(np.exp(-dt*np.sum(r[:int(T1/dt),:], axis=0))*discounted_payoff)
        price += caplet_price
    return price

print("Cap price (risk neutral measure): {:.4f}".format(cap_price_risk_neutral(t, T, delta, K, r_rn)))
```

#### 12. Conduct the valuation under the **T-forward measure**.

```python
def mc_sim_T_forward_measure(T, N, r0, a, sigma, theta, NT):
    r = np.zeros((NT+1,N))
    r[0,:] = r0
    for i in range(NT):
        eps = np.random.normal(0,1,N)
        r[i+1,:] = r[i,:]+(theta(i*dt)-a*r[i,:]-sigma**2*Phi(i*dt,T))*dt+sigma*np.sqrt(dt)*eps
    return r

def cap_price_T_forward_measure(t, T, delta, K, r):
    price=0
    for ti in np.arange(t, T, delta):
        T1 = ti
        T2 = ti+delta
        P_T1_T2 = P(T1,T2,r)
        libor = (1-P_T1_T2)/(delta*P_T1_T2)
        payoff = np.maximum(libor-K,0)*delta
        caplet_price = np.mean(P(t,T2,r)*payoff)
        price += caplet_price
    return price

r_t_fwd = mc_sim_T_forward_measure(T, N, r0, a, sigma, theta, NT)
print("Cap price (T forward measure): {:.4f}".format(cap_price_T_forward_measure(t, T, delta, K, r_t_fwd)))
```

We obtain almost the same price, which is coherent with the fact that the price is independent of the chosen measure.

#### 13. Determine a closed-form value for the cap. Recall that the price of a European put option with strike $K$ and maturity $T$, written on a zero-coupon bond with maturity $S$ at time $t \in[0, T]$, is given by:
$$
Z B P(t, T, S, K)=-P(t, S) \Phi(-h)+K P(t, T) \Phi(-h+\tilde{\sigma}) \tag{12}
$$
#### with:
$$
\tilde{\sigma}=\sigma \sqrt{\frac{1-e^{-2 a(T-t)}}{2 a}} B(T, S) \tag{13}
$$
$$
B(T, S)=\frac{1-e^{-a(S-T)}}{a} \tag{14}
$$
$$
h=\frac{1}{\tilde{\sigma}} \ln \left(\frac{P(t, S)}{P(t, T) K}\right)+\frac{\tilde{\sigma}}{2} \tag{15}
$$

```python
def ZBP(t, T, K, S, r):
    return -P(t, S,r)[0]*sc.stats.norm.cdf(-h(t,T,S,K,r)) + K*P(t, T,r)[0]*sc.stats.norm.cdf(-h(t,T,S,K,r)+sigma_tilde(t, T, S))
def sigma_tilde(t, T, S):
    return sigma*np.sqrt((1-np.exp(-2*a*(T-t)))/(2*a))*B(T,S)
def B(T,S):
    return (1-np.exp(-a*(S-T)))/a
def h(t,T,S,K,r):
    return (1/sigma_tilde(t, T, S))*np.log(P(t, S, r)[0]/(P(t, T, r)[0]*K)) + sigma_tilde(t, T, S)/2

def cap_price_formula(t, T, delta, S, K, r):
    return (1+K*delta)*ZBP(t, T, 1/(1+K*delta), S, r)

print("Cap price (from put formula): {:.4f}".format(cap_price_formula(t, t, delta, T, K, r_rn)))
```

```python
def caplet_price_formula(t, T, delta, S, K, r):
    return (1+delta*K)*ZBP(t, T, 1/(1+delta*K), S, r)

def cap_price_zcb(t, T, delta, K, r):
    price=0
    for ti in np.arange(t, T, delta):
        caplet_price = caplet_price_formula(t, ti, delta, ti+delta, K, r)
        price += caplet_price
    return price

print("Cap price (sum of caplets): {:.4f}".format(cap_price_zcb(t, T, delta, K, r_rn)))
```

#### 14. Propose a Monte Carlo simulation for the pricing of the European Swaption.
We use our original risk-neutral Monte Carlo simulation.

```python
def get_swap_rate_formula(delta, t, n, tau, beta0, beta1, beta, bonds):
    payments= np.linspace(t, t+n*delta, n+1)
    alpha=0
    beta=payments.shape[0]-1
    S = (bonds[alpha] - bonds[beta]) / (delta*np.sum(bonds[alpha+1:beta+1,:], axis=0))
    return S
```

```python
def european_swaption(delta, t, T, K, r, tau, beta0, beta1, beta2):
    n=int(T/delta)
    payments=np.linspace(t, T, n+1)
    bond_prices = np.array([P(t, payment, r) for payment in payments])
    swap_rate = get_swap_rate_formula(delta, t, n, tau, beta0, beta1, beta2, bond_prices)
    payoff = np.maximum(swap_rate - K, 0) * np.sum(bond_prices, axis=0) * delta
    discount = np.exp(-np.sum(r, axis=0)*dt)
    price = np.mean(discount*payoff)
    return price
print("Swaption price: {:.4f}".format(european_swaption(delta, t, T, K, r_rn, tau, beta0, beta1, beta2)))
```

#### 15. Propose a Monte Carlo simulation for the pricing of the Bermudan Swaption.


```python
def bermudan_swaption(delta, t, T, K, r, tau, beta0, beta1, beta2, exercise_dates):
    n = int(T / delta)
    payments = np.linspace(t, T, n + 1)
    bond_prices = np.array([P(t, payment, r) for payment in payments])

    # Initialize the value of the swaption at maturity
    value = np.maximum(get_swap_rate_formula(delta, t, n, tau, beta0, beta1, beta2, bond_prices) - K, 0) * np.sum(bond_prices, axis=0) * delta

    for i, exercise_date in enumerate(reversed(exercise_dates[:-1])):
        idx = int((exercise_date - t) / delta)
        bond_prices = np.array([P(t, payment, r[idx:, :]) for payment in payments[idx:]])
        swap_rate = get_swap_rate_formula(delta, exercise_date, n - idx, tau, beta0, beta1, beta2, bond_prices)
        exercise_payoff = np.maximum(swap_rate - K, 0) * np.sum(bond_prices, axis=0) * delta
        if i == 0:
            value = exercise_payoff
        else:
            continuation_value = value * np.exp(-dt*np.sum(r[idx:, :], axis=0))
            value = np.maximum(exercise_payoff, continuation_value)
        
    # Discount the final value back to the present
    price = np.mean(value)
    return price

exercise_dates = np.arange(1, T + 1)  # Example exercise dates

price = bermudan_swaption(delta, t, T, K, r_rn, tau, beta0, beta1, beta2, exercise_dates)
print("Bermudan Swaption price: {:.4f}".format(price))
```

We find a price that is higher than the European price, which is coherent. 

#### 16. Why is the correlation among interest rates important in the valuation of swaptions?

The price of a swaption depends on the expected future paths of interest rates. Contrary to caps, where we can value each caplet independently because there is no correlation between payoffs, the swap rate is on the whole period and thus presents correlation between rates at different points in time. 

Correlation between interest rates at different maturities influences these paths, as highly correlated rates will tend to move together and thus produce only parallel shifts to the current yield curve. We also need a lower correlation to take into account the potential change of shape of the yield curve. 

Also, a high correlation would lead to a high volatility of the swap rate (that depends on forward rates) because all forward rates would move in the same way, leading to a potentially overpriced swaption. A lower correlation would lead to a lower volatility of the swap rate because moves of forward rates may offset each other. 

#### 17. Explain why the 1-factor HW model does not capture the correlation among interest rate models.
The 1-factor HW model is driven by a single stochastic factor, which means all interest rates are perfectly correlated. This single factor determines the movement of the entire yield curve, implying parallel shifts of the yield curve, which is not the case in reality because different economic situations lead to different shapes of the yield curve. To address this issue, multi-factor models have been introduced.
